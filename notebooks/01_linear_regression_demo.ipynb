# Linear Regression from Scratch

This notebook demonstrates a NumPy-based implementation of linear regression using gradient descent.

The objective is to understand how a model learns the relationship between input features and target values by minimizing the mean squared error.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt

from ml_scratch.linear_model.linear_regression import LinearRegression
```

## 1. Generate Synthetic Data

We create a simple linear dataset with random noise.

The data roughly follows the equation:

y = 4 + 3x + noise

```python
np.random.seed(42)

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X.squeeze() + np.random.randn(100)

plt.figure(figsize=(8, 5))
plt.scatter(X, y)
plt.title("Synthetic Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

## 2. Model Overview

Linear regression assumes a linear relationship between input and output:

y = wx + b

Where:
- w is the weight
- b is the bias

The model is trained by minimizing the Mean Squared Error (MSE):

MSE = (1/n) * Σ(y - ŷ)^2

To optimize the parameters, gradient descent is used.

```python
model = LinearRegression(lr=0.05, n_iters=1000)
model.fit(X, y)
```

## 3. Prediction Result

After training, the model predicts a regression line that fits the data.

```python
y_pred = model.predict(X)

sorted_idx = np.argsort(X[:, 0])
X_sorted = X[sorted_idx]
y_pred_sorted = y_pred[sorted_idx]

plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Data")
plt.plot(X_sorted, y_pred_sorted, color="red", label="Prediction")
plt.title("Linear Regression Result")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 4. Learned Parameters

```python
print("Learned weight:", model.weights)
print("Learned bias:", model.bias)
```

## 5. Loss Curve

The loss curve shows how the mean squared error decreases during training.

```python
plt.figure(figsize=(8, 5))
plt.plot(model.loss_history)
plt.title("Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("MSE Loss")
plt.show()
```

## 6. Conclusion

- The model successfully learned the linear trend in the data.
- The loss decreased over iterations, which indicates that gradient descent worked properly.
- Linear regression is simple but effective when the relationship between variables is approximately linear.
